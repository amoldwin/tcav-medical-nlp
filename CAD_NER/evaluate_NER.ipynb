```python
import tensorflow as tf
import pandas as pd
from transformers import BertTokenizerFast, BatchEncoding, TFBertForTokenClassification
import numpy as np
from sklearn.metrics import recall_score, classification_report
```

```python
tokenizer = BertTokenizerFast.from_pretrained("emilyalsentzer/Bio_ClinicalBERT")
```

```python
val_df = pd.read_pickle('datasets/n2c2_val_dataset.pk')
```

```python
val_inputs_dict = {'input_ids':np.array([x for x in val_df['chunked_tokens']])}
```

```python
val_dataset = tf.data.Dataset.from_tensor_slices((
    val_inputs_dict,
    list(val_df['label_ids'])
))
```

```python
model = TFBertForTokenClassification.from_pretrained('NER/models/clinicalbert_batchsize8_complete_brat_test90.h5')
```

```python
outputs=model.predict(val_dataset.batch(8))
```

```python
predictions = tf.argmax(outputs[0], axis=2)
```

```python
prediction_probabilities = tf.reduce_max(outputs[0], axis=2)
```

```python
labels_list =['B-CAD', 'B-DIABETES', 'B-FAMILY_HIST', 'B-HYPERLIPIDEMIA', 'B-HYPERTENSION', 'B-MEDICATION', 'B-OBESE', 'B-PHI', 'B-SMOKER', 'I-CAD', 'I-DIABETES', 'I-FAMILY_HIST', 'I-HYPERLIPIDEMIA', 'I-HYPERTENSION', 'I-MEDICATION', 'I-OBESE', 'I-PHI', 'I-SMOKER', 'O']
```

```python
reverse_labels_dict = {i:k for i,k in enumerate(labels_list)}
```

```python
val_results_df = pd.DataFrame([list(predictions.numpy())]).transpose()
```

```python
val_results_df['tokens'] = list(val_df['chunked_tokens'].apply(lambda x: [tokenizer.decode([y]) for y in x]) )
```

```python
val_results_df.columns=['class_nums','tokens']
```

```python
val_results_df['pred_probs'] = list(prediction_probabilities.numpy())
```

```python
val_results_df['classes']=val_results_df['class_nums'].apply(lambda x: [reverse_labels_dict[y] for y in x] )
```

```python
val_y = val_df['label_ids']
```

```python
val_results_df['token_labels'] = pd.Series(list(val_y)).apply(lambda x: [reverse_labels_dict[y] for y in x])
```

```python
val_results_df['filename'] = list(val_df['filename'])
val_results_df['evidence_spans'] = list(val_df['evidence_spans'])
val_results_df['encoded'] = list(val_df['encoded'])
val_results_df['chunked_idxs'] = list(val_df['chunked_idxs'])
val_results_df['note_text'] = list(val_df['note_text'])
val_results_df['labels'] = list(val_df['labels'])
```

```python
def doc_level_stats(labels, preds):
    doc_preds = [[label in preds[i] for label in labels_list] for i in range(len(preds))]
    doc_labels = [[label in labels[i] for label in labels_list] for i in range(len(preds))]
    print(classification_report(doc_labels, doc_preds, target_names=labels_list))
```

```python
def get_token_spans(row, col):
    label_lst=row[col]
    label_lst=[x[2:]  if '-' in x else x for x in label_lst]
    start = 0
    end =0
    spans=[]
    labels=[]
    char_spans=[]
    for i in range(1,len(label_lst)):
        if label_lst[i]!=label_lst[i-1]:
            end=i-1
            spans.append((start,end,label_lst[i]))
            labels.append(label_lst[i-1])
            start=i
        if i==len(label_lst)-1:
            end=i
            spans.append((start,end,label_lst[i]))
            labels.append(label_lst[i])
    return spans

def get_label_run_spans(row, col):
    try:
        first_pad_idx = row['encoded']['input_ids'].index(0)-1
    except: 
        first_pad_idx=len(row['chunked_idxs'])
    label_lst=row[col][1:first_pad_idx]
    label_lst=[x[2:]  if '-' in x else x for x in label_lst]
    start = 0
    end =0
    spans=[]
    labels=[]
    char_spans=[]
    for i in range(1,len(label_lst)):
        if label_lst[i]!=label_lst[i-1]:
            end=i-1
            spans.append((start,end,label_lst[i]))
            labels.append(label_lst[i-1])
            start=i
        if i==len(label_lst)-1:
            end=i
            spans.append((start,end,label_lst[i]))
            labels.append(label_lst[i])
    for i, span in enumerate(spans):
        try:
            start = row['encoded'].token_to_chars(row['chunked_idxs'][span[0]]).start
            end = row['encoded'].token_to_chars(row['chunked_idxs'][span[1]]).end
            char_spans.append((labels[i],start,end))
        except:
            print(first_pad_idx,row['tokens'][span[1]:])
            print(row['encoded']['input_ids'])
            
    
    return spans,char_spans,labels
```
