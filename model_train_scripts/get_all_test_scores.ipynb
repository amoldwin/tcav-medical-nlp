```python
import pandas as pd
import os
```

```python
# for fn in os.listdir('checkpoints/final_scores'):
#     print(fn)
scores_dir = 'checkpoints/final_scores'
```

```python
fns = {}
fns['NICU_logistic'] = 'FINALLogistic_0_metrics.csv'
fns['NICU_CNN'] = 'snapshot_2022-01-25_14-41-17_metrics.csv'
fns['NICU_BERT']='FINALWeightedBlueBert512NicuLosAdmissions_2.csv'

fns['METABOLIC_logistic']='FINALLogisticCCS_halfsplit_0FINALLogisticCCS_halfsplit_0_metrics.csv'
fns['METABOLIC_CNN']='METABOLIC_halfsplit_2022-01-26_10-00-55_metrics.csv'
fns['METABOLIC_BERT']='CCS_BERT_TEXT__0.csv'
fns['METABOLIC_bert_residual'] = 'residual_CCS__0.csv'
fns['METABOLIC_residual_CNN'] = 'METABOLIC_halfsplit_2022-01-26_09-48-15_metrics.csv'
fns['METABOLIC_residual_just_structured'] = 'CCS_just_residual_0.csv'
```

```python
def isfloat(value):
  try:
    float(value)
    return True
  except ValueError:
    return False
```

```python
all_scores = pd.DataFrame(columns=['model','val_mcc','test_accuracy', 'test_mcc', #'test_auc_avg', 'test_auprc_avg',
       'test_auc_1', 'test_auprc_1', 'test_precision', 'test_recall',
       'test_f1-score', 'test_support', 'test_specificity'])
for k in fns.keys():
    df = pd.read_csv(os.path.join(scores_dir,fns[k]))
    df.columns = df.columns.str.lower()
    df = df[df['val_mcc'].apply(lambda x: isfloat(str(x)))]
    df['val_mcc'] =pd.to_numeric(df["val_mcc"], downcast="float")
    df = df[df['val_mcc']==df['val_mcc'].max() ].iloc[:1]
    print(df['val_mcc'].max(), type(df['val_mcc'].max()))
    df = df[[col for col in df.columns if 'val_mcc' in col or ( 'test' in col and not 'avg' in col) ]]
    df=df[[ 'val_mcc', 'test_accuracy', 'test_mcc', 'test_auc_1', 'test_auprc_1', 'test_precision', 'test_recall', 'test_f1-score','test_support', 'test_specificity']]
    print(df.columns)
    df['model']=k
    all_scores = pd.concat([all_scores, df])
```

```python
all_scores.reset_index(drop=True).applymap(lambda x: float(x) if isfloat(x) else x ).round(3)
```
